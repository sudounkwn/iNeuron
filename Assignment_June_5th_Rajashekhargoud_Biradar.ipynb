```python
##2. Try to write a function which  is equivalent  to print function in python

import sys
def printfunc(x):
    l=len(x)
    for i in range(l):
        sys.stdout.write(x[i])

printfunc(input('enter the content to be printed: '))
```

```python
##3. Try to write a function which is a replica of list append , extend and pop function

li1=[1,2,3,4] 
li2=[5,6,7,8]

def listappend(x,y):
    '''Function to that resembles inbuilt list append operation'''
    l=len(x)
    x.insert(l,y)
    return x

listappend(li1,li2)

def listextend(x,y):
    '''Function to that resembles inbuilt list extend operation'''
    l=len(x) #4
    n=len(y) #4
    j=0
    for c in range(n):
        x.insert(l,y[j])
        j+=1
        l+=1
    return x
listextend(li1,li2)


def listpop(x):
    '''Function to that resembles inbuilt list pop operation'''
    i=int(input('enter the index of element you want to remove--Index starting from 0'))
    l=len(x)
    z=[]
    for j in range(l):
        if i==j:
            pass
        else:
            z.append(x[j])
    return z

listpop(li1)
```
